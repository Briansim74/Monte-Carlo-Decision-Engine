```python
import random

# Full pre-flop hand ranges by position
preflop_ranges = {
    "early": {
        # Pocket pairs
        "AA": 1.0, "KK": 1.0, "QQ": 1.0, "JJ": 1.0,
        "TT": 0.3, "99": 0.2, "88": 0.1, "77": 0.0,
        "66": 0.0, "55": 0.0, "44": 0.0, "33": 0.0, "22": 0.0,

        # Suited broadways
        "AKs": 1.0, "AQs": 1.0, "AJs": 0.8, "ATs": 0.3,
        "KQs": 1.0, "KJs": 0.5, "QJs": 0.5,

        # Offsuit broadways
        "AKo": 1.0, "AQo": 1.0, "AJo": 0.8, "KQo": 1.0, "KJo": 0.3, "QJo": 0.2,

        # Suited connectors & one-gappers
        "JTs": 0.2, "T9s": 0.2, "98s": 0.1, "87s": 0.1, "76s": 0.0, "65s": 0.0, "54s": 0.0,

        # Suited aces
        "A2s": 0.0, "A3s": 0.0, "A4s": 0.0, "A5s": 0.0
    },
    "middle": {
        # Pocket pairs
        "AA": 1.0, "KK": 1.0, "QQ": 1.0, "JJ": 1.0,
        "TT": 0.8, "99": 0.6, "88": 0.5, "77": 0.4,
        "66": 0.3, "55": 0.2, "44": 0.1, "33": 0.0, "22": 0.0,

        # Suited broadways
        "AKs": 1.0, "AQs": 1.0, "AJs": 1.0, "ATs": 0.8, "A9s": 0.5, "A8s": 0.3,
        "KQs": 1.0, "KJs": 0.8, "KTs": 0.5, "QJs": 0.8, "QTs": 0.5,

        # Offsuit broadways
        "AKo": 1.0, "AQo": 1.0, "AJo": 0.8, "ATo": 0.5,
        "KQo": 1.0, "KJo": 0.5, "QJo": 0.5, "QTo": 0.3,

        # Suited connectors & one-gappers
        "JTs": 0.5, "T9s": 0.5, "98s": 0.3, "87s": 0.3, "76s": 0.2, "65s": 0.1, "54s": 0.1,
        "64s": 0.1, "75s": 0.1, "86s": 0.1, "97s": 0.1,

        # Suited aces
        "A2s": 0.1, "A3s": 0.1, "A4s": 0.1, "A5s": 0.1
    },
    "late": {
        # Pocket pairs
        "AA": 1.0, "KK": 1.0, "QQ": 1.0, "JJ": 1.0,
        "TT": 1.0, "99": 1.0, "88": 1.0, "77": 1.0,
        "66": 0.8, "55": 0.8, "44": 0.7, "33": 0.7, "22": 0.7,

        # Suited broadways
        "AKs": 1.0, "AQs": 1.0, "AJs": 1.0, "ATs": 1.0, "A9s": 0.8, "A8s": 0.7, "A7s": 0.6,
        "A6s": 0.5, "A5s": 0.5, "A4s": 0.5, "A3s": 0.5, "A2s": 0.5,
        "KQs": 1.0, "KJs": 1.0, "KTs": 0.8, "K9s": 0.6, "QJs": 1.0, "QTs": 0.8, "Q9s": 0.6,

        # Offsuit broadways
        "AKo": 1.0, "AQo": 1.0, "AJo": 0.8, "ATo": 0.8,
        "KQo": 1.0, "KJo": 0.8, "KTo": 0.5, "QJo": 0.8, "QTo": 0.5,

        # Suited connectors & one-gappers
        "JTs": 0.8, "T9s": 0.8, "98s": 0.6, "87s": 0.6, "76s": 0.5, "65s": 0.5, "54s": 0.5,
        "64s": 0.4, "75s": 0.4, "86s": 0.4, "97s": 0.4,

        # Suited aces
        "A2s": 0.5, "A3s": 0.5, "A4s": 0.5, "A5s": 0.5
    }
}


def get_positions_order(num_players):
    """
    Generate poker seat names for a table of num_players.
    Seats are ordered clockwise from UTG (first to act pre-flop after BB).
    Uses standard naming:
        Early: UTG, UTG+1, UTG+2...
        Middle: MP, MP+1...
        Late: CO, BTN
        Blinds: SB, BB
    """

    if num_players == 2:
        return ["SB", "BB"]
    elif num_players == 3:
        return ["UTG", "SB", "BB"]
    elif num_players == 4:
        return ["UTG", "MP", "SB", "BB"]
    elif num_players == 5:
        return ["UTG", "MP", "CO", "SB", "BB"]
    elif num_players == 6:
        return ["UTG", "MP", "CO", "BTN", "SB", "BB"]
    elif num_players == 7:
        return ["UTG", "UTG+1", "MP", "CO", "BTN", "SB", "BB"]
    elif num_players == 8:
        return ["UTG", "UTG+1", "UTG+2", "MP", "CO", "BTN", "SB", "BB"]
    elif num_players == 9:
        return ["UTG", "UTG+1", "UTG+2", "MP", "MP+1", "CO", "BTN", "SB", "BB"]
    else:
        raise ValueError("Supports 2-9 players only")


def get_general_position(player_index, positions_order):
    pos_name = positions_order[player_index]

    if pos_name.startswith("UTG") or pos_name == "SB" or pos_name == "BB":
        return "early"
    
    elif pos_name.startswith("MP"):
        return "middle"
    
    else:
        return "late"
    

def adjust_weight_for_previous_raises(weight, num_raises):
    """
    Adjust hand weight dynamically based on previous raises.
    """
    if num_raises == 0:
        return weight
    elif num_raises == 1:
        return weight * 0.5  # reduce medium/weak hands
    else:
        return max(0, weight - 0.3)  # strong raises only play strong hands


def deal_poker_hand(num_players=6):
    """
    Creates a shuffled deck, deals 2 cards per player,
    converts them to poker hand notation (AKs, AKo, AA),
    and deals flop, turn, river with burn cards.
    """

    # Build deck
    ranks = "23456789TJQKA"
    suits = "hdcs"

    deck = [r + s for r in ranks for s in suits]
    random.shuffle(deck)

    def cards_to_hand(card1, card2):
        r1, s1 = card1[0], card1[1]
        r2, s2 = card2[0], card2[1]

        # Ensure higher rank comes first
        if ranks.index(r1) < ranks.index(r2):
            r1, r2 = r2, r1
            s1, s2 = s2, s1

        # Pocket pair
        if r1 == r2:
            return r1 + r2

        suited = "s" if s1 == s2 else "o"
        return r1 + r2 + suited

    player_hands = {}

    # Deal hole cards
    for i in range(num_players):
        c1 = deck.pop()
        c2 = deck.pop()

        player_hands[i] = {
            "cards": [c1, c2],
            "hand": cards_to_hand(c1, c2)
        }

    # ---- FLOP ----
    burn1 = deck.pop()
    flop = [deck.pop(), deck.pop(), deck.pop()]

    # ---- TURN ----
    burn2 = deck.pop()
    turn = flop + [deck.pop()]

    # ---- RIVER ----
    burn3 = deck.pop()
    river = turn + [deck.pop()]

    return player_hands, flop, turn, river


def rotate_seating(seating):
    print("Rotating seating...")
    
    #new_seating = seating[1:] + seating[:1]
    new_seating = seating[-1:] + seating[:-1]
    print("New Seating:", new_seating)

    return new_seating


def initialize_players_and_stack(num_players=6, stack=1000):
    return [{"id": i, "stack": stack} for i in range(num_players)]
```

### Preflop

```python
def decide_preflop_action(hand, position, pot_size, actions_so_far, stack, contribution, current_bet, last_raise_size=0):
    """
    Decide pre-flop action using hand, position, pot, previous actions, and stack.
    Enforces minimum raise rules.
    """
    weight = preflop_ranges[position].get(hand, 0)
    num_raises = sum(1 for a in actions_so_far if a["action"] == "raise")
    weight = adjust_weight_for_previous_raises(weight, num_raises)

    if weight < 0.3:
        return "fold", 0

    action_roll = random.random()
    call_amount = max(0, current_bet - contribution) # no need for our simulation engine

    # Determine raise amount respecting min-raise
    def get_raise_amount(extra_multiplier):
        # intended_raise = extra chips you *want to add* on top of calling
        intended_raise = call_amount + pot_size * extra_multiplier

        # min_total_raise = minimum total contribution required to make a legal raise
        min_total_raise = current_bet + last_raise_size

        # total_raise = total contribution you will have in the pot after your raise
        # ensures raise respects the minimum
        total_raise = max(intended_raise + contribution, min_total_raise)

        # actual chips to put in now, capped by your stack, allows for 'all in' without raising to the required legal amount
        return min(stack, total_raise - contribution)

    if weight < 0.7:
        if num_raises == 0:
            if action_roll < 0.6:
                bet = get_raise_amount(0.5 + weight)
                return "raise", bet
            
            else:
                return "call", min(stack, call_amount)
        
        else:
            return "call", min(stack, call_amount)
    
    else:
        if num_raises == 0:
            bet = get_raise_amount(1 + weight)
            return "raise", bet
        
        elif num_raises == 1:
            bet = get_raise_amount(1.5 + weight)
            return "raise", bet
        
        else:
            if weight > 0.9:
                bet = get_raise_amount(2 + weight)
                return "raise", bet
            
            else:
                return "fold", 0
```

### Initialize Preflop

```python
def initialize_preflop(game_state, players, my_player_id, sb_amount=10):
    
    seating = game_state["seating"] if game_state is not None else list(range(len(players)))
    print("Current Seating:", seating)

    print("\nDEALING CARDS...")

    print("\nPRE-FLOP\n")

    num_players = len(players)

    game_state = {
        "pot": 0,
        "actions": [],
        "current_bet": 0,
        "last_raise_size": 0,
        "seating": seating,
        "players": players,
        "starting_stacks": {p["id"]: p["stack"] for p in players},
        "player_state": {
            p["id"]: {
                "hand": None,
                "cards": None,
                "contribution": 0,
                "folded": False
            }
            for p in players
        },
        "my_player_id": my_player_id,
        "game_ended": False
    }


    # Deal cards
    player_hands, flop, turn, river = deal_poker_hand(num_players=num_players)
    game_state["board"] = {
        "flop": flop,
        "turn": turn,
        "river": river
    }

    for seat_index in range(num_players):
        player_id = game_state["seating"][seat_index]

        game_state["player_state"][player_id]["hand"] = player_hands[seat_index]["hand"]
        game_state["player_state"][player_id]["cards"] = player_hands[seat_index]["cards"]


    positions_order = get_positions_order(num_players)

    # Blinds based on seat index
    sb_index = (num_players - 2)
    bb_index = (num_players - 1)
    bb_amount = 2 * sb_amount

    # ---- POST BLINDS ----
    blinds = [
        (sb_index, sb_amount, "SB"),
        (bb_index, bb_amount, "BB")
    ]

    for seat_idx, amt, name in blinds:
        player_id = game_state["seating"][seat_idx]
        player_state = game_state["player_state"][player_id]
        position = get_general_position(seat_idx, positions_order)

        bet = min(amt, game_state["players"][player_id]["stack"])
        game_state["players"][player_id]["stack"] -= bet

        player_state["contribution"] += bet
        game_state["pot"] += bet
        game_state["current_bet"] = max(game_state["current_bet"], player_state["contribution"])

        game_state["actions"].append({
            "player_index": player_id,
            "hand": player_state["hand"],
            "position": position,
            "action": f"post {name}",
            "bet": bet,
            "street": "Pre-Flop"
        })

        print(f"Player {player_id}: ({positions_order[seat_idx]}, {position}) with {player_state["hand"]} -> post {name} {bet:.2f}{" (YOU)" if player_id == game_state["my_player_id"] else ""}")


    game_state["last_raise_size"] = bb_amount - sb_amount

    action_index = (bb_index + 1) % num_players
    players_to_act = num_players


    return game_state, positions_order, action_index, players_to_act, num_players
```

### Hand Strength Evaluation

```python
from collections import Counter
from itertools import combinations

def evaluate_hand_strength(cards, board):

    ranks = "23456789TJQKA"
    rank_values = {r:i for i,r in enumerate(ranks, start=2)}

    all_cards = cards + board

    ranks_list = [c[0] for c in all_cards]
    suits_list = [c[1] for c in all_cards]

    rank_counts = Counter(ranks_list)
    suit_counts = Counter(suits_list)

    values = sorted([rank_values[r] for r in ranks_list])

    # ---------- FLUSH ----------
    flush = any(count >= 5 for count in suit_counts.values())

    # ---------- STRAIGHT ----------
    unique_vals = sorted(set(values))

    straight = False
    for i in range(len(unique_vals)-4):
        if unique_vals[i+4] - unique_vals[i] == 4:
            straight = True

    # Wheel straight (A2345)
    if set([14,2,3,4,5]).issubset(unique_vals):
        straight = True

    # ---------- MULTIPLES ----------
    counts = sorted(rank_counts.values(), reverse=True)

    four = 4 in counts
    three = 3 in counts
    pairs = counts.count(2)

    # ---------- HAND TYPE ----------
    if straight and flush:
        score = 0.98
    elif four:
        score = 0.95
    elif three and pairs >= 1:
        score = 0.90
    elif flush:
        score = 0.85
    elif straight:
        score = 0.80
    elif three:
        score = 0.70
    elif pairs >= 2:
        score = 0.60
    elif pairs == 1:
        score = 0.50
    else:
        # High card strength scaling
        high = max(values)
        score = 0.2 + (high / 14) * 0.2

    return score



def evaluate_hand_strength_combo(cards, board):

    ranks = "23456789TJQKA"
    rank_values = {r:i for i,r in enumerate(ranks, start=2)}

    HAND_RANKS = {
        "High Card": 0,
        "One Pair": 1,
        "Two Pair": 2,
        "Three of a Kind": 3,
        "Straight": 4,
        "Flush": 5,
        "Full House": 6,
        "Four of a Kind": 7,
        "Straight Flush": 8,
    }

    all_cards = cards + board
    all_cards = sorted(all_cards, key=lambda c: rank_values[c[0]], reverse=True)

    ranks_list = [c[0] for c in all_cards]
    suits_list = [c[1] for c in all_cards]

    rank_counts = Counter(ranks_list)
    suit_counts = Counter(suits_list)

    values = sorted([rank_values[r] for r in ranks_list], reverse=True)

    # Helper: get cards of a rank
    def get_cards_by_value(val, count):
        return [c for c in all_cards if rank_values[c[0]] == val][:count]

    # ---------- FLUSH ----------
    flush_suit = None
    for suit, count in suit_counts.items():
        if count >= 5:
            flush_suit = suit

    flush_cards = []
    if flush_suit:
        flush_cards = [c for c in all_cards if c[1] == flush_suit][:5]

    # ---------- STRAIGHT ----------
    unique_vals = sorted(set(values))
    straight_high = None
    for i in range(len(unique_vals) - 4):
        if unique_vals[i+4] - unique_vals[i] == 4:
            straight_high = unique_vals[i+4]

    # Wheel (A-2-3-4-5)
    if set([14,2,3,4,5]).issubset(unique_vals):
        straight_high = 5

    def get_straight_cards(high):
        needed = [high - i for i in range(5)]
        result = []
        for v in needed:
            for c in all_cards:
                if rank_values[c[0]] == v and c not in result:
                    result.append(c)
                    break
        return result

    # ---------- MULTIPLES ----------
    counts = sorted(rank_counts.items(), key=lambda x: (-x[1], -rank_values[x[0]]))
    counts = [(rank_values[r], c) for r,c in counts]

    # ---------- HAND TYPE CHECKS ----------

    # Straight Flush
    if flush_suit and straight_high:
        sf_cards = [c for c in all_cards if c[1] == flush_suit]
        sf_values = sorted(set([rank_values[c[0]] for c in sf_cards]))
        for i in range(len(sf_values) - 4):
            if sf_values[i+4] - sf_values[i] == 4:
                high = sf_values[i+4]
                combo = get_straight_cards(high)
                tiebreak = [rank_values[c[0]] for c in combo]
                return (HAND_RANKS["Straight Flush"], tiebreak), "Straight Flush", combo
        if set([14,2,3,4,5]).issubset(sf_values):
            combo = get_straight_cards(5)
            tiebreak = [rank_values[c[0]] for c in combo]
            return (HAND_RANKS["Straight Flush"], tiebreak), "Straight Flush", combo

    # Four of a Kind
    if counts[0][1] == 4:
        quad = get_cards_by_value(counts[0][0], 4)
        kicker = [c for c in all_cards if rank_values[c[0]] != counts[0][0]][0]
        combo = quad + [kicker]
        tiebreak = [rank_values[c[0]] for c in combo]
        return (HAND_RANKS["Four of a Kind"], tiebreak), "Four of a Kind", combo

    # Full House
    if counts[0][1] == 3 and counts[1][1] >= 2:
        trips = get_cards_by_value(counts[0][0], 3)
        pair = get_cards_by_value(counts[1][0], 2)
        combo = trips + pair
        tiebreak = [rank_values[c[0]] for c in combo]
        return (HAND_RANKS["Full House"], tiebreak), "Full House", combo

    # Flush
    if flush_suit:
        combo = flush_cards
        tiebreak = [rank_values[c[0]] for c in combo]
        return (HAND_RANKS["Flush"], tiebreak), "Flush", combo

    # Straight
    if straight_high:
        combo = get_straight_cards(straight_high)
        tiebreak = [rank_values[c[0]] for c in combo]
        return (HAND_RANKS["Straight"], tiebreak), "Straight", combo

    # Three of a Kind
    if counts[0][1] == 3:
        trips = get_cards_by_value(counts[0][0], 3)
        kickers = [c for c in all_cards if rank_values[c[0]] != counts[0][0]][:2]
        combo = trips + kickers
        tiebreak = [rank_values[c[0]] for c in combo]
        return (HAND_RANKS["Three of a Kind"], tiebreak), "Three of a Kind", combo

    # Two Pair
    if counts[0][1] == 2 and counts[1][1] == 2:
        pair1 = get_cards_by_value(counts[0][0], 2)
        pair2 = get_cards_by_value(counts[1][0], 2)
        kicker = [c for c in all_cards if rank_values[c[0]] not in (counts[0][0], counts[1][0])][0]
        combo = pair1 + pair2 + [kicker]
        tiebreak = [rank_values[c[0]] for c in combo]
        return (HAND_RANKS["Two Pair"], tiebreak), "Two Pair", combo

    # One Pair
    if counts[0][1] == 2:
        pair = get_cards_by_value(counts[0][0], 2)
        kickers = [c for c in all_cards if rank_values[c[0]] != counts[0][0]][:3]
        combo = pair + kickers
        tiebreak = [rank_values[c[0]] for c in combo]
        return (HAND_RANKS["One Pair"], tiebreak), "One Pair", combo

    # High Card
    combo = all_cards[:5]
    tiebreak = [rank_values[c[0]] for c in combo]
    return (HAND_RANKS["High Card"], tiebreak), "High Card", combo
```

```python
# Cactus Kev / 7-card evaluator

ranks = "23456789TJQKA"
suits = "hdcs"
rank_values = {r:i for i,r in enumerate(ranks, start=2)}

def evaluate_hand_strength_7cards(cards, board):
    """
    Simple 7-card evaluator inspired by Cactus Kev.
    Returns a numeric strength. Higher is stronger.
    """
    all_cards = cards + board
    best_score = 0
    
    # Evaluate all 5-card combinations
    for combo in combinations(all_cards, 5): # Outputs 7C5 combos from the board and your hand
        score = evaluate_5card(combo)
        if score > best_score:
            best_score = score
    return best_score

def evaluate_5card(cards):
    """
    Returns a numeric score for a 5-card hand.
    Higher = stronger.
    """
    ranks_list = [c[0] for c in cards]
    suits_list = [c[1] for c in cards]
    
    # Counts
    rank_counts = Counter(ranks_list)
    suit_counts = Counter(suits_list)
    values = sorted([rank_values[r] for r in ranks_list], reverse=True)
    
    # Check flush
    flush = max(suit_counts.values()) == 5
    
    # Check straight
    unique_vals = sorted(set(values))
    straight = False
    if len(unique_vals) == 5 and unique_vals[-1] - unique_vals[0] == 4:
        straight = True
    # Wheel straight
    if set([14,2,3,4,5]).issubset(values):
        straight = True
    
    counts = sorted(rank_counts.values(), reverse=True)
    four = 4 in counts
    three = 3 in counts
    pairs = counts.count(2)
    
    # Hand rank scoring
    if straight and flush:
        return 8000000 + max(values)  # Straight flush
    elif four:
        return 7000000 + max(values)  # Four of a kind
    elif three and pairs >= 1:
        return 6000000 + max(values)  # Full house
    elif flush:
        return 5000000 + max(values)  # Flush
    elif straight:
        return 4000000 + max(values)  # Straight
    elif three:
        return 3000000 + max(values)  # Three of a kind
    elif pairs >= 2:
        return 2000000 + max(values)  # Two pair
    elif pairs == 1:
        return 1000000 + max(values)  # One pair
    else:
        return max(values)  # High card
```

### Monte Carlo Range Equity

<div style="font-size:13px;">
Instead of guessing hand strength, we simulate thousands of random runouts.

Example process:

Hero: AhKh
Board: Qh 7d 2c
Opponent: random hand

Steps repeated many times:

Generate opponent hand

Deal remaining turn/river

Evaluate both hands

Count wins / losses

After many simulations:

equity = wins / total

Example on a flop:

Hero equity: 0.62

Meaning 62% chance to win vs random hand.
</div>

```python
def monte_carlo_equity_multiway(hero_cards, board, num_opponents, simulations=2000):
    """
    Estimate hero equity against multiple active opponents using Monte Carlo.
    hero_cards: list of 2 cards
    board: list of flop/turn/river cards
    active_hands: list of opponent hands (2 cards each)
    """

    wins = 0
    ties = 0

    # Build deck
    ranks = "23456789TJQKA"
    suits = "hdcs"

    base_deck = [r+s for r in ranks for s in suits if r+s not in hero_cards + board]

    for _ in range(simulations):
        deck = base_deck[:]   # ✅ fresh shallow deck copy every simimulation
        random.shuffle(deck)

        # Deal opp hands
        opp_hands = [[deck.pop(), deck.pop()] for _ in range(num_opponents)]

        # Complete board
        remaining_board = board + [deck.pop() for _ in range(5 - len(board))]

        # Evaluate
        # hero_strength = evaluate_hand_strength(hero_cards, remaining_board) # DEFAULT COUNTER
        # opp_strengths = [evaluate_hand_strength(h, remaining_board) for h in opp_hands]

        hero_strength = evaluate_hand_strength_7cards(hero_cards, remaining_board) # 7 WAY COUNTER  # CAN TRY TO USE THIS INSTEAD OR COMBO HAND STRENGTHS
        opp_strengths = [evaluate_hand_strength_7cards(h, remaining_board) for h in opp_hands]

        if all(hero_strength > o for o in opp_strengths):
            wins += 1
        elif any(hero_strength == o for o in opp_strengths):
            ties += 1

    return (wins + ties * 0.5) / simulations
```

previous_actions
</div>

```python
def decide_flop_action(hero_cards, board, num_opponents, pot, stack, contribution, current_bet, last_raise_size):
    """
    Make flop decisions based on equity vs multiple opponents and board texture.
    """

    board_texture = analyze_board_texture(board)
    equity = monte_carlo_equity_multiway(hero_cards, board, num_opponents)

    call_amount = max(0, current_bet - contribution)
    action_roll = random.random()

    def get_raise_amount(multiplier):
        intended_raise = call_amount + pot * multiplier
        min_total_raise = current_bet + last_raise_size
        total_raise = max(intended_raise + contribution, min_total_raise)
        return min(stack, total_raise - contribution)

    # Equity-based thresholds
    if equity > 0.65:
        # Strong hands → bet or raise
        if current_bet == contribution:
            bet_multiplier = 0.7 if board_texture["flush_draw"] else 0.6
            return "bet", min(stack, pot * bet_multiplier)
        
        else:
            return "raise", get_raise_amount(1.2 if board_texture["flush_draw"] else 1.0)

    elif equity > 0.35:
        # Medium strength → call or small raise
        if current_bet == contribution:
            if board_texture["paired"]:
                return "check", 0
            
            elif action_roll < 0.4:
                return "bet", min(stack, pot * 0.5)
            
            else:
                return "check", 0
            
        else:
            if action_roll < 0.8:
                return "call", min(stack, call_amount)
            
            else:
                return "raise", get_raise_amount(0.7)

    else:
        # Weak hands → fold or occasional bluff
        if current_bet == contribution:
            bluff_chance = 0.15 + (0.2 if board_texture["flush_draw"] else 0)

            if action_roll < bluff_chance:
                return "bet", min(stack, pot * 0.5)
            
            else:
                return "check", 0
            
        else:
            fold_chance = 0.7 + (0.2 if board_texture["paired"] else 0)

            if action_roll < fold_chance:
                return "fold", 0
            
            else:
                return "call", min(stack, call_amount)
```

### Turn

<div style="font-size:15px;">
Turn should: <br>

reduce bluff frequency

increase aggression with strong hands

allow semi-bluffs if board is wet
</div>

```python
def decide_turn_action(hero_cards, board, num_opponents, pot, stack, contribution, current_bet, last_raise_size):
    """
    Make turn decisions based on equity vs multiple opponents and board texture.
    
    Turn strategy:
    - Lower bluff frequency
    - Higher aggression with strong hands
    - Semi-bluffing on wet boards
    """

    board_texture = analyze_board_texture(board)
    equity = monte_carlo_equity_multiway(hero_cards, board, num_opponents)

    call_amount = max(0, current_bet - contribution)
    action_roll = random.random()

    def get_raise_amount(multiplier):
        intended_raise = call_amount + pot * multiplier
        min_total_raise = current_bet + last_raise_size
        total_raise = max(intended_raise + contribution, min_total_raise)
        return min(stack, total_raise - contribution)

    # --- Detect draws (very important for semi-bluffs) ---
    suits = [c[1] for c in board] + [hero_cards[0][1], hero_cards[1][1]]
    suit_counts = Counter(suits)

    has_flush_draw = any(count == 4 for count in suit_counts.values())

    # Simple straight draw detection
    ranks = "23456789TJQKA"
    rank_values = sorted(set([ranks.index(c[0]) for c in board + hero_cards]))
    
    has_straight_draw = any(
        rank_values[i+3] - rank_values[i] <= 4
        for i in range(len(rank_values) - 3)
    )

    semi_bluff = has_flush_draw or has_straight_draw

    # --- STRONG HANDS ---
    if equity > 0.7:
        if current_bet == contribution:
            # Bigger bets on turn
            return "bet", min(stack, pot * (0.8 if board_texture["flush_draw"] else 0.7))
        
        else:
            # More aggressive raises
            return "raise", get_raise_amount(1.3 if board_texture["flush_draw"] else 1.1)

    # --- MEDIUM HANDS ---
    elif equity > 0.4:
        if current_bet == contribution:
            if action_roll < 0.3:
                return "bet", min(stack, pot * 0.5)
            
            else:
                return "check", 0
            
        else:
            if action_roll < 0.75:
                return "call", min(stack, call_amount)
            
            else:
                return "fold", 0

    # --- WEAK HANDS / SEMI-BLUFF ---
    else:
        if current_bet == contribution:
            # Only semi-bluff, not pure bluff
            if semi_bluff and board_texture["flush_draw"]:
                if action_roll < 0.4:
                    return "bet", min(stack, pot * 0.6)
                
            return "check", 0
        
        else:
            if semi_bluff:
                # Continue with draws sometimes
                if action_roll < 0.5:
                    return "call", min(stack, call_amount)
                
                elif action_roll < 0.7:
                    return "raise", get_raise_amount(0.8)
                
            return "fold", 0
```

### River

<div style="font-size:15px;">
River should: <br>

almost no bluffs

strong hands value bet

medium hands mostly call

weak hands fold
</div>

```python
def decide_river_action(hero_cards, board, num_opponents, pot, stack, contribution, current_bet, last_raise_size):
    """
    Make river decisions based on equity and board texture.

    River strategy:
    - Almost no bluffs
    - Strong hands → value bet / raise
    - Medium hands → mostly call
    - Weak hands → fold
    """

    board_texture = analyze_board_texture(board)
    equity = monte_carlo_equity_multiway(hero_cards, board, num_opponents)

    call_amount = max(0, current_bet - contribution)
    action_roll = random.random()

    def get_raise_amount(multiplier):
        intended_raise = call_amount + pot * multiplier
        min_total_raise = current_bet + last_raise_size
        total_raise = max(intended_raise + contribution, min_total_raise)
        return min(stack, total_raise - contribution)

    # --- STRONG HANDS ---
    if equity > 0.75:
        if current_bet == contribution:
            # Value bet
            return "bet", min(stack, pot * (0.8 if board_texture["flush_draw"] else 0.7))
        
        else:
            # Raise for value
            return "raise", get_raise_amount(1.2 if board_texture["flush_draw"] else 1.0)

    # --- MEDIUM HANDS ---
    elif equity > 0.4:
        if current_bet == contribution:
            # Mostly check, occasionally bet small
            if action_roll < 0.2:
                return "bet", min(stack, pot * 0.4)
            
            else:
                return "check", 0
        else:
            # Mostly call
            return "call", min(stack, call_amount)

    # --- WEAK HANDS ---
    else:
        if current_bet == contribution:
            return "check", 0
        
        else:
            return "fold", 0
```

### Street Simulation

```python
def decide_postflop_action(street, hero_cards, board, num_opponents, pot, stack, contribution, current_bet, last_raize_size):

    if street == "flop":
        action, bet = decide_flop_action(hero_cards, board, num_opponents, pot, stack, contribution, current_bet, last_raize_size)

    elif street == "turn":
        action, bet = decide_turn_action(hero_cards, board, num_opponents, pot, stack, contribution, current_bet, last_raize_size)

    elif street == "river":
        action, bet = decide_river_action(hero_cards, board, num_opponents, pot, stack, contribution, current_bet, last_raize_size)

    return action, bet
```

```python
def simulate_street(game_state, street, players=None, my_player_id=0, sb_amount=10):

    # Initialize preflop game_state
    if street == "preflop":
        game_state, positions_order, action_index, players_to_act, num_players = initialize_preflop(game_state, players, my_player_id=my_player_id, sb_amount=sb_amount)
    
    # Initialize postflop game_state
    else:
        if game_state["game_ended"] == True:
            return game_state
    
        game_state, board, positions_order, action_index, players_to_act, num_players = initialize_postflop(game_state, street)


    while players_to_act > 0:
        player_id = game_state["seating"][action_index]
        player_state = game_state["player_state"][player_id]

        # Skip folded players
        if not player_state["folded"]:
            hand = player_state["hand"]
            position = get_general_position(action_index, positions_order)
            hero_cards = game_state["player_state"][player_id]["cards"]

            active_players = sum(1 for player_id, s in game_state["player_state"].items() if not s["folded"])
            num_opponents = active_players - 1  # subtract hero

            # preflop
            if street == "preflop":
                action, bet = decide_preflop_action(
                    hand,
                    position,
                    game_state["pot"],
                    game_state["actions"],
                    game_state["players"][player_id]["stack"],
                    player_state["contribution"],
                    game_state["current_bet"],
                    game_state["last_raise_size"]
                )

            # flop, turn, river
            else:
                action, bet = decide_postflop_action(
                    street,
                    hero_cards,
                    board,
                    num_opponents,
                    game_state["pot"],
                    game_state["players"][player_id]["stack"],
                    player_state["contribution"],
                    game_state["current_bet"],
                    game_state["last_raise_size"]
                )
   
            # Apply action
            if action == "fold":
                player_state["folded"] = True
            

            elif action == "call":
                game_state["players"][player_id]["stack"] -= bet
                player_state["contribution"] += bet
                game_state["pot"] += bet


            else: # call / bet / raise / check
                game_state["players"][player_id]["stack"] -= bet
                player_state["contribution"] += bet
                game_state["pot"] += bet

                # Detect new bet / raise
                new_total = player_state["contribution"]
                if new_total > game_state["current_bet"]:
                    game_state["last_raise_size"] = new_total - game_state["current_bet"]
                    game_state["current_bet"] = new_total
                    # Reset action cycle
                    players_to_act = num_players


            # Record action
            game_state["actions"].append({
                "player_index": player_id,
                "hand": hand,
                "position": position,
                "action": action,
                "bet": bet,
                "street": f"{street.capitalize()}"
            })

            print(f"Player {player_id}: ({positions_order[action_index]}, {position}) with {player_state["hand"]} -> {action} {bet:.2f}{" (YOU)" if player_id == game_state["my_player_id"] else ""}")

        # Move to next player
        action_index = (action_index + 1) % num_players
        players_to_act -= 1

        # Skip folded players
        while player_state["folded"] and players_to_act > 0:
            player_id = game_state["seating"][action_index]
            player_state = game_state["player_state"][player_id]
            
            if player_state["folded"]:
                action_index = (action_index + 1) % num_players
                players_to_act -= 1
                continue


        # End hand if only 1 player remains
        active_players = sum(1 for player_id, s in game_state["player_state"].items() if not s["folded"])
        if active_players == 1:
            winner_id = next(player_id for player_id, s in game_state["player_state"].items() if not s["folded"])
            position = get_general_position(game_state["seating"].index(winner_id), positions_order)
            
            # Update Pot
            game_state["players"][winner_id]["stack"] += game_state["pot"]

            game_state["actions"].append({
                "player_index": winner_id,
                "hand": game_state["player_state"][winner_id]["hand"],
                "position": position,
                "action": "win_uncontested",
                "bet": game_state["pot"],
                "street": f"{street.capitalize()}"
            })

            game_state["game_ended"] = True

            print(f"Player {winner_id}: ({positions_order[game_state["seating"].index(winner_id)]}, {position}) with {game_state["player_state"][winner_id]["hand"]} -> win_uncontested {game_state["pot"]:.2f}{" (YOU)" if winner_id == game_state["my_player_id"] else ""}")
            print(f"Player {winner_id}: ({positions_order[game_state["seating"].index(winner_id)]}, {position}) wins uncontested {game_state["pot"]:.2f} {"during " if street != "flop" else "post"}{street}! {"(YOU)" if winner_id == game_state["my_player_id"] else ""}")

            print("\nEND OF GAME")
            print(f"Final pot size: {game_state['pot']:.2f}")

            # NET PROFIT
            net_profits = {}
            for p in game_state["players"]:
                start = game_state["starting_stacks"][p["id"]]
                net_profits[p["id"]] = {
                    "net_profit": p["stack"] - start,
                    "stack": p["stack"]
                }

            print("\nUpdated Stacks and Net Profits:")
            for player_id, stats in net_profits.items():
                print(f"Player {player_id}: {stats["stack"]} ({'+' if stats["net_profit"] > 0 else ''}{stats["net_profit"]})")

            print(f"\nHand ended {street}.\n")

            # Rotate seating
            game_state["seating"] = rotate_seating(game_state["seating"])

            return game_state


    return game_state
```

### Showdown Simulation

```python
def simulate_showdown(game_state):
    """
    Resolve the final showdown at the river for remaining active players.
    """

    if game_state["game_ended"] == True:
        return game_state
    
    print("\nSHOWDOWN\n")

    board = game_state["board"]["river"]
    print(f"Board - river:", board, "\n")

    num_players = len(game_state["starting_stacks"])
    positions_order = get_positions_order(num_players)

    active_indices = [i for i, s in game_state["player_state"].items() if not s["folded"]]

    hand_strengths = {}
    for player_id in active_indices:
        player_state = game_state["player_state"][player_id]

        hero_cards = player_state["cards"]
        rank, hand_type, combo = evaluate_hand_strength_combo(hero_cards, board)

        hand_strengths[player_id] = {
                    "rank": rank, # tuple (hand_rank, tiebreak) for perfect comparison
                    "hand_type": hand_type,
                    "combo": combo
                }

        position = get_general_position(player_id, positions_order)

        print(f"Player {player_id}: ({positions_order[player_id]}, {position}) with {player_state["hand"]} -> cards: ({player_state["cards"][0]}, {player_state["cards"][1]}), hand_type: {hand_type}, combo: {combo}{" (YOU)" if player_id == game_state["my_player_id"] else ""}")
    
    
    # Determine winners
    max_rank = max(s["rank"] for s in hand_strengths.values())

    winners = [
        player_id
        for player_id, s in hand_strengths.items()
        if s["rank"] == max_rank
    ]

    # Split the pot among winners
    pot_share = game_state["pot"] / len(winners)

    for winner_id in winners:
        position = get_general_position(winner_id, positions_order)

        game_state["players"][winner_id]["stack"] += pot_share

        game_state["actions"].append({
            "player_index": winner_id,
            "hand": game_state["player_state"][winner_id]["hand"],
            "position": position,
            "action": "win_showdown",
            "bet": game_state["pot"],
            "street": "River",
        })
        
        print(f"Player {winner_id}: ({positions_order[winner_id]}, {position}) with {game_state["player_state"][winner_id]["hand"]} -> win_showdown {"(YOU)" if winner_id == game_state["my_player_id"] else ""}")
        print(f"Player {winner_id}: ({positions_order[winner_id]}, {position}) wins {pot_share:.2f} from the showdown!")


    print("\nEND OF GAME")
    print(f"Final pot size: {game_state['pot']:.2f}")

    
    # NET PROFIT
    net_profits = {}
    for p in game_state["players"]:
        start = game_state["starting_stacks"][p["id"]]
        net_profits[p["id"]] = {
            "net_profit": p["stack"] - start,
            "stack": p["stack"]
        }

    print("\nUpdated Stacks and Net Profits:")
    for player_id, stats in net_profits.items():
        print(f"Player {player_id}: {stats["stack"]} ({'+' if stats["net_profit"] > 0 else ''}{stats["net_profit"]})")

    print("\nHand ended showdown.\n")

    # Rotate seating
    game_state["seating"] = rotate_seating(game_state["seating"])

    return game_state
```

### Run Full Simulation

```python
def run_simulation(game_state, players, num_players, my_player_id, sb_amount):
    print("my_player_id", my_player_id)
    # Preflop
    game_state = simulate_street(game_state=game_state, street="preflop", players=players, my_player_id=my_player_id, sb_amount=sb_amount)

    # Postflop
    game_state = simulate_street(game_state=game_state, street="flop")
    
    # Turn
    game_state = simulate_street(game_state, "turn")
    
    # River
    game_state = simulate_street(game_state, "river")
    
    # Showdown
    game_state = simulate_showdown(game_state)
    
    return game_state
```

```python
def run_simulations(stack=1000, num_players=6, my_player_id=0, sb_amount=10, simulations=5):
    
    # Initialize players
    players = initialize_players_and_stack(num_players=num_players, stack=stack)
    print("\n--------- RUNNING SIMULATIONS ---------\n")
    print("--------- GAME 1 ---------\n")

    # Initialize game
    game_state = run_simulation(game_state=None, players=players, num_players=num_players, my_player_id=my_player_id, sb_amount=sb_amount)

    for i in range(2, simulations + 1):
        print(f"\n\n--------- GAME {i} ---------\n")

        game_state = run_simulation(game_state=game_state, players=game_state["players"], num_players=num_players, my_player_id=my_player_id, sb_amount=sb_amount)

    print("\n--------- END OF SIMULATIONS ---------")


run_simulations(stack=1000, num_players=6, my_player_id=1, sb_amount=10, simulations=3)
```
